# Task 1: Data Preparation and Missing Value Handling (30 points)

```python
import pandas as pd

print("=== TASK 1: Data Preparation ===\n")

# Create Students DataFrame
students_data = {
    "student_id": [101, 102, 103, 104, 105, 106, 107],
    "name": ["Alice", "Bob", None, "David", "Emma", "Frank", "Grace"],
    "email": ["alice@email.com", "bob@email.com", None,
              "david@email.com", "emma@email.com",
              "frank@email.com", "grace@email.com"],
    "city": ["Mumbai", "Delhi", "Chennai", None,
             "Pune", "Hyderabad", "Bangalore"]
}

students_df = pd.DataFrame(students_data)

print("Original Students DataFrame:")
print(students_df)
print()

# Null Analysis
print("Null Value Analysis:")
for col in students_df.columns:
    null_count = students_df[col].isnull().sum()
    percent = (null_count / len(students_df)) * 100
    print(f"Column: {col}, Nulls: {null_count} ({percent:.2f}%)")

# Cleaning
students_df["city"] = students_df["city"].fillna("Unknown")
students_df.dropna(subset=["name"], inplace=True)

print(f"\nCleaned Students DataFrame ({len(students_df)} rows):")
print(students_df)
```

# Task 2: Multiple Join Operations (40 points)


```python
print("\n=== TASK 2: Join Operations ===\n")

# Enrollments DataFrame
enrollments_data = {
    "student_id": [101, 102, 105, 103, 108, 109],
    "course_name": ["Python", "Data Science", "Machine Learning",
                    "AI", "Cloud Computing", "Cyber Security"],
    "enrollment_date": ["2024-01-15", "2024-01-20", "2024-02-10",
                        "2024-02-18", "2024-03-01", "2024-03-05"]
}

enrollments_df = pd.DataFrame(enrollments_data)

# INNER JOIN
inner_join = pd.merge(students_df, enrollments_df, on="student_id", how="inner")
print(f"Inner Join Result ({len(inner_join)} rows):")
print(inner_join[["student_id","name","course_name","enrollment_date"]])
print("Excluded students:", list(set(students_df["student_id"]) - set(inner_join["student_id"])),
      "- Not in enrollments table\n")

# LEFT JOIN
left_join = pd.merge(students_df, enrollments_df, on="student_id", how="left")
print(f"Left Join Result ({len(left_join)} rows):")
print(left_join[["student_id","name","course_name","enrollment_date"]])
null_students = left_join[left_join["course_name"].isnull()]["student_id"].tolist()
print("Students with null course_name:", null_students, "\n")

# RIGHT JOIN
right_join = pd.merge(students_df, enrollments_df, on="student_id", how="right")
print(f"Right Join Result ({len(right_join)} rows):")
print(right_join[["student_id","name","course_name","enrollment_date"]], "\n")

# OUTER JOIN
outer_join = pd.merge(students_df, enrollments_df, on="student_id", how="outer", indicator=True)
print(f"Outer Join Result ({len(outer_join)} rows):")
print(outer_join)
```

# Task 3: Lookup Operation and Automation (30 points)


```python
print("\n=== TASK 3: Lookup and Automation ===\n")

# Exam Scores
scores_data = {
    "student_id": [101, 102, 104, 105, 106],
    "exam_score": [85, 92, 78, 88, 95]
}

scores_df = pd.DataFrame(scores_data)

lookup_result = pd.merge(students_df, scores_df, on="student_id", how="left")

print("Lookup Operation Result:")
print(lookup_result[["student_id","name","exam_score"]])
print()

# Automation Function
def join_tables(join_type):
    result = pd.merge(students_df, enrollments_df, on="student_id", how=join_type)
    print("Join Type:", join_type)
    print("Rows in Result:", len(result))
    print("Result Preview:")
    print(result[["student_id","name","course_name"]])
    print()

print("Automation Function Test:")
join_tables("inner")

print("Left Join Test:")
join_tables("left")
```
